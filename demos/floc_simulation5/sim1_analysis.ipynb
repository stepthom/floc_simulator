```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('sim1_out.csv')
```

```python
df
```

```python
res = df.groupby(['d', 'per_add'])['match'].describe().reset_index()
res
```

```python
res[['d', 'per_add', 'count', 'mean']]
```

```python
#sns.set_style("whitegrid")
#sns.set(rc={'figure.figsize':(11.7,8.27)})
#sns.set(font_scale = 2)
sns.set_context("paper")
sns.set_style("whitegrid")

ax = sns.lineplot(x="per_add", y="mean", hue="d", data=res)
ax.set(xlabel="Percentage of New Domains")
ax.set(ylabel="Probality of Same Cohort ID")
ax.legend(title='Starting domains')
plt.tight_layout()
```
