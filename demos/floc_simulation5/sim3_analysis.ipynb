```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import ast
```

```python
df = pd.read_csv('out/sim3_user_weeks.csv')
```

```python
df['cohort_ids'] = df['cohort_ids'].apply(lambda x: ast.literal_eval(x))
```

```python
for weeks in [int(i) for i in np.arange(1, 53, step=1, dtype=int)]:
    df['hash_{}'.format(weeks)] = df['cohort_ids'].apply(lambda x: hash(tuple(x[0:weeks])))
    num = df['hash_{}'.format(weeks)].nunique()
    per = 100 * (num / df.shape[0])
    print("First {:2d} weeks: {:5d} / {:6d} = {:.1f}% unique".format(weeks, num, df.shape[0], per ))
```
